```
# Tugas Pengganti Presensi Pertemuan 12
# Laurentius Nathaniel (Python G8) ( 1101194377 )
import numpy as np

# menentukan c1 dengan fr = Frekuensi (MHz)
def c1(fr) : 
    if fr >= 400 and fr <= 1500 :
        return 69.55
    elif fr > 1500 and fr <= 2000 :
        return 46.3

# menentukan c2 dengan fr = Frekuensi (MHz)
def c2(fr) : 
    if fr >= 400 and fr <= 1500 :
        return 26.16
    elif fr > 1500 and fr <= 2000 :
        return 33.9

# Mencari ahR dengan hR = Tinggi Pemancar dan fr = Frekuensi
def a(hR, fr) : 
    return (((1.1*np.log10(fr))-0.7)*hR) - ((1.56*np.log10(fr))-0.8)

# Mencari Cm dengan Type_Cm = Tipe Area dan fr = Frekuensi (MHz)
def Cm(Type_Cm, fr) : 
    if Type_Cm == 1 :
        return 0
    elif Type_Cm == 2 :
        return (-2*((np.log10(fr/28))**2))-5.4
    else :
        return (-4.78*((np.log10(fr))**2))+(18.33*np.log10(fr))-40.94

# Fungsi Pathloss
def Pathloss(d, fr, Type_Cm, hT, hR) :
  return c1(fr)+(c2(fr)*np.log10(fr))-(13.83*np.log10(hT))-a(hR, fr)+((44.9-(6.55*np.log10(hT)))*np.log10(d))+Cm(Type_Cm, fr)
```

```
# Bagian Input Parameter

print("Masukkan parameter! \n")

d = float(input("Jarak Pemancar (Tx) ke Penerima (Rx) (km) : "))

fr = float(input("Frekuensi (400 MHz - 2000 MHz) : "))
while fr < 400 or fr > 2000 : # selama diluar rentang frekuensi
 print("Frekuensi Diluar Rentang !!")
 fr = float(input("Frekuensi (400 MHz - 2000 MHz) : "))


print("""Tipe Area :\n
1. Urban\n
2. Suburban\n
3. Open\n""")
Type_Cm = int(input("Tipe Area (1/2/3) : "))
#selama pilihan diluar tipe
while Type_Cm < 1 and Type_Cm >3 :
 print("Anda belum memilih Tipe Area !!")
 print("""Tipe Area :\n
 1. Urban\n
 2. Suburban\n
 3. Open\n""")
 Type_Cm = int(input("Tipe Area (1/2/3) : "))
 
if Type_Cm == 1 :
  Tipe_Area = "Urban"
elif Type_Cm == 2 :
  Tipe_Area = "Suburban"
else :
  Tipe_Area = "Open"


hT = float(input("Tinggi Pemancar (30 m - 200 m) : "))
while hT < 30 or hT > 200 : # selama diluar rentang
 print("Tinggi Pemancar Diluar Rentang !!")
 hT = float(input("Tinggi Pemancar (30 m - 200 m) : "))

hR = float(input("Tinggi Penerima (1 m - 30 m) : "))
while hR < 1 or hR > 30 : # selama diluar rentang
 print("Tinggi Penerima Diluar Rentang !!")
 hR = float(input("Tinggi Penerima (1 m - 30 m) : "))

print("\n\nParameter yang dimasukkan :\n")
print("""1. Jarak Pemancar ke Penerima  : {} km\n
2. Frekuensi  : {} MHz\n
3. Tipe Area  : Area {}\n
4. Tinggi Pemancar  : {} m\n
5. Tinggi Penerima  : {} m\n""".format(d, fr, Tipe_Area, hT, hR))

# Pemanggilan fungsi untuk output pathloss
Lp = Pathloss(d, fr, Type_Cm, hT, hR)
print("\nPathloss berdasarkan parameter diatas adalah :", Lp , "dB")
```
